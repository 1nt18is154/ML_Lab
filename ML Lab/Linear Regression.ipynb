```python
import pandas as pd   
import statistics as st 
import math
import scipy.stats
import matplotlib.pyplot as plt
data=pd.read_csv('data.csv')
print(data)

sx=st.stdev(data['x'])
print("Standard Derivation of x",sx)

sy=st.stdev(data['y'])
print("Standard Derivation of y",sy)

r=scipy.stats.pearsonr(data['x'],data['y'])[0]
print("Correlation coefficient ",r)

#slope
slope=r*sy/sx
print("Slope",slope)

mx=st.mean(data['x'])
my=st.mean(data['y'])
print(mx,my)

c=my-slope*mx
print("c",c)

y_pred=slope*data['x']+c
print(y_pred)
#for i in range(len(data)):
#    y1.append(slope*data['x'][i]+c)
#print(y1)

#plotting graph
plt.scatter(data['x'],data['y'])
plt.plot(data['x'],y_pred,color="red")

#cost
cost=abs(data['y']-y_pred)
print("Cost is ",cost.values)

#sum squared error
sse=sum(cost**2)
print("Sum of Square error ",sse)

```
